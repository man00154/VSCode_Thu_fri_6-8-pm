1. Univariate analysis

```python
df['species'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['species'].value_counts().plot(kind='bar',
                                  title='valuecounts',
                                  figsize=(8, 6))
                               plt.show()
```

```python
for i in cat:
    df[i].value_counts().plot(kind='bar',title='count plot for df[i]')
```

Continuous features : Histogram

```python
con
```

```python
for i in con:
    sns.histplot(data=df, x=i, kde=True)
    plt.title(f'Histogram for {i}')
    plt.show()
```

```python
con
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width')
plt.title('Sepal Length vs Sepal Width')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='petal_length', y='petal_width')
plt.title(f'Scatterplot for petal length vs petal width')
plt.grid()
plt.show()
```

Plot the scatter plot for

```python
sns.heatmap(df[con].corr(),annot=True, fmt='.3f')
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Boxplot for species vs Sepal Length')
plt.show()
```

```python
for i in con:
    plt.figure(figsize=(16,8))
    sns.boxplot(data=df,x='species',y=i)
    plt.title(f'Boxplot of Species vs {i}')
    plt.show()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset
data = pd.read_csv('infy_stock.csv')

# Assuming the dataset has 'Date' and 'Close' columns
dates = pd.to_datetime(data['Date'])
close_prices = data['Close']

# Perform time series decomposition
decomposition = seasonal_decompose(close_prices, model='additive', period=30)  # Assuming a seasonal period of 30 days

# Extract decomposed components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot results
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(dates, close_prices, label='Original')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(dates, trend, label='Trend')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(dates, seasonal, label='Seasonal')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(dates, residual, label='Residual')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

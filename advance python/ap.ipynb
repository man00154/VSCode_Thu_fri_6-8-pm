```python
def welcome_decorator(func) :
    def wrapper(*args,**kwargs) :
        print('hi user')
        print(f'function result : {func(*args, **kwargs)}')
        return func(*args, **kwargs)
    return wrapper


```

```python
@welcome_decorator
def mul (a,b) :
    c = a*b
    return c
```

```python
mul(3,4)
```

```python
def authenticate(func):
    # For illustration purpose i created users dictionary with
    # Key contains usernames and Value contains password
    user = {'utkarshg1':'Example@123',
            'aditim4':'ETLHive#901',
            'sarthaks2':'Security_456',
            'ramans3':'Example&789'}
    def wrapper(username, password, *args, **kwargs):
        if username in user.keys() and user.get(username)==password:
            print(f'Welcome {username} Authentication successfull')            
            return func(username, *args, **kwargs)
        else:
            print('Authentication FAILED!')
            print('Please check username and password.')
    return wrapper
```

```python
@authenticate
def compound_intrest(username, P, N, R):
    """
    P : Prinicipal in INR,
    N : Period in Years
    R : Rate of Intrest in %p.a.
    """
    A = P*(1+R/100)**N
    I = A-P
    return I, A
```

```python
class Account:
    
    #Constructor function - To define attributes
    def __init__(self, ac_no, name, bal):
        self.ac_no = ac_no
        self.name = name
        self.bal = bal
        
    #Create a deposit function
    def deposit(self, amt):
        self.bal = self.bal + amt
        print(f'Amount of {amt} INR Deposited in account no : {self.ac_no}')
        print(f'Updated balance : {self.bal} INR')
        
    #Withdraw function
    def withdraw(self, amt):
        if amt<=self.bal:
            self.bal = self.bal - amt
            print(f'Amount of {amt} INR Withdrawn from your account no : {self.ac_no}')
            print(f'Updated Balance : {self.bal} INR')
        else:
            print('Transaction declined because of Insufficent Balance')
            print(f'Current Balance : {self.bal} INR')
            
    #Transfer to another account
    def transfer(self, amt, ac2):
        if amt<=self.bal:
            self.bal = self.bal - amt
            ac2.bal = ac2.bal + amt
            print(f'Amount of {amt} INR transferred from your ac no : {self.ac_no} to {ac2.ac_no}')
            print(f'Updated Balance : {self.bal} INR')
        else:
            print('Transaction declined because of Insufficent Balance')
            print(f'Current Balance : {self.bal} INR')         
```

```python
def divide2():
    try:
        num = float(input('Please enter numerator   : '))
        den = float(input('Please enter denominator : '))
        d = num/den
        return d
    except ZeroDivisionError:
        print('Denominator cannot be Zero.')
    except ValueError:
        print('Please enter numbers only inside numerator and denominator')
    except Exception as e:
        print(f'Unexpected error occured : {e}')
```

```python
def file_download(url):
    filename = url.split('/')[-1]
    print(f'Downloading File : {filename}')
    urlretrieve(url, filename)
    print(f'{filename} Download Complete')
    print('\n=======================================\n')
```

```python
url
```

```python
%%time
file_download(url)
```

```python
urls = ["https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/data.csv",
        "https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/test.csv",
        "https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/train.csv"]
```

```python
%%time
for i in urls:
    file_download(i)
```

```python
from threading import Thread
```

```python
%%time
# Create and start threads
threads = []
for i in urls:
    th = Thread(target=file_download, args=(i,))
    th.start()
    threads.append(th)

# Wait for all threads to finish
for thr in threads:
    thr.join()
```

```python
from bs4 import BeautifulSoup
bs = BeautifulSoup(data)
```

Finding Product names

```python
product_tags = bs.find_all('div', class_="tUxRFH")
product_tags
```

```python
product_tags[0]
```

```python
product_tags[0].text
```

```python
product_names = []
for i in product_tags:    
    product_names.append(i.text)
```

```python
product_names
```

```python
price_tags = bs.find_all('div', class_="Nx9bqj _4b5DiR")
price_tags
```

```python
prices = []
for i in price_tags:
    p = i.text[1:].replace(',','')
    p = float(p)
    prices.append(p)
```

```python
prices
```

```python
dct = {'product_names':product_names,
      'Price':prices,
      }
```

```python
import pandas as pd
df = pd.DataFrame(dct)
df
```
